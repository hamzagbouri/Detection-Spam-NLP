```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("test").getOrCreate()
```

```python
df = spark.read.csv(r"C:\Users\hamza\Desktop\Systeme-Predictif-Gestion-Logistique\data\data_final_v2\data.csv", inferSchema=True, header=True)
```

```python
df.show(5)
```

```python
from pyspark.sql.functions import col, avg, count
df.filter(col("late_delivery_risk") == 1).groupBy("order_region").agg(
    avg("sales").alias("Moyenne Salaire"),
    count("late_delivery_risk").alias("Total Risk")
    ).show()
```
